## Despliegue

La interfaz web es accesible desde https://www.ezsuarez.org/Prediccion/#/dashboard

La interfaz web permite seleccionar automáticamente un ejemplo de entrada válido, visualizar y editar el JSON correspondiente y enviarlo al modelo para obtener una predicción. El resultado se muestra junto con la probabilidad de riesgo.

El modelo de Machine Learning fue desplegado sobre una Raspberry Pi, exponiéndose como un servicio REST mediante FastAPI. El modelo serializado (`.pkl`) se carga al iniciar la aplicación, permitiendo realizar predicciones sin necesidad de reentrenamiento.

La API se ejecuta como un servicio del sistema utilizando **systemd**, garantizando su disponibilidad permanente y reinicio automático ante fallos. Para el acceso externo, se configuró **Apache** como *reverse proxy*, redirigiendo las solicitudes HTTP hacia la API interna sin exponer directamente el backend ni los archivos del modelo.

El servicio quedó disponible de forma segura a través de una solicitud POST de HTTPS en el siguiente endpoint: https://www.ezsuarez.org/api/predict

El estado del servicio puede ser consultado a través de una solicitud GET de HTTPS en el siguiente endpoint: http://www.ezsuarez.org/api/health

Las pruebas de funcionamiento se realizaron mediante Postman, validando la correcta recepción de datos de entrada, la ejecución del modelo y la entrega de predicciones con sus respectivas probabilidades.

En la carpeta **artefactos** se incluyen los siguientes elementos:
- **`app.py`**: aplicación FastAPI encargada de exponer los endpoints de predicción y de salud, integrada al servidor Apache mediante configuración de *reverse proxy*.
- **Servicio systemd**: archivo de configuración que permite ejecutar la API como un servicio del sistema, asegurando su inicio automático y disponibilidad continua.
- **Modelo entrenado (`.pkl`)**: artefacto del modelo de Machine Learning previamente entrenado y cargado al iniciar la aplicación.
- **Archivo `.txt` con JSON de prueba**: contiene ejemplos de solicitudes utilizadas para validar el correcto funcionamiento del endpoint de predicción.
